```python
# In NumPy, the retstep parameter in the np.linspace function allows you to get both the array of values and the spacing between values (step size).

# The formula for calculating the step size when retstep=True is:

# step=(stop−start)/(num−1)

 


# start is the starting value of the sequence,
# stop is the end value of the sequence,
# num is the total number of samples.
# When you set retstep=True, np.linspace returns a tuple: the array of values and the step size.print(np.linspace(1,100))
print(np.linspace(1,100))
print(np.linspace(1,100,5))
print(np.linspace(1,100,10))
print(np.linspace(1,100,10,endpoint=False))
print(np.linspace(1,100,10,endpoint=False,retstep=True))
print(np.linspace(1,100,retstep=True))
print(np.linspace(1,100,endpoint=False))
```

```python
#first dataframe
d={'id':[1,2,3,4],
  'val':['a','b','c','d']}
a={'id':[1,2,8,14],
  'val':['p','q','r','s']}
df1=pd.DataFrame(d)
df2=pd.DataFrame(a)


#inner join returns a dataframe having same/common  parameters
df3=pd.merge(df1,df2,on='id', how='inner')
print(df3)
```

```python
d={'id':[1,2,3,4],
  'val':['a','b','c','d']}
a={'id':[1,2,8,14],
  'val':['p','q','r','s']}
df1=pd.DataFrame(d)
df2=pd.DataFrame(a)



df3=pd.merge(df1,df2,on='id', how='left')
print(df3)
```
